# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score

best_rstate=0
accu=0

for i in range(0,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=i)
    mod=LinearRegression()
    mod.fit(x_train,y_train)
    y_pred=mod.predict(x_test)
    tempaccu=r2_score(y_test,y_pred)
    
    if tempaccu>accu:
        accu=tempaccu
        best_rstate=i
        
print('Best accuracy is ',(accu*100),'found on random state',(best_rstate))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=190)
LR=LinearRegression()
LR.fit(x_train,y_train)
y_pred=LR.predict(x_test)
score=r2_score(y_test,y_pred)
cvscore=cross_val_score(LR,x_train,y_train,cv=5).mean()

print(f"r2 score is {score*100} and cross validation score is {cvscore*100}")
```

# Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, classification_report
Rf=RandomForestClassifier()
Rf.fit(x_train,y_train)
p=Rf.predict(x_test)
print(f"accuracy score is {accuracy_score(y_test,p)*100}")
print("cvscore",cross_val_score(Rf,x,y,cv=5).mean())
print('\n Classification Report:\n', classification_report(y_test,p))
```

# Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, classification_report
Lr=LogisticRegression()
Lr.fit(x_train,y_train)
p=Lr.predict(x_test)
print(f"accuracy score is {accuracy_score(y_test,p)*100}")
print("cvscore",cross_val_score(Lr,x,y,cv=5).mean())
print('\n Classification Report:\n', classification_report(y_test,p))
```

# SVM Classifier

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, recall_score, classification_report
svc_rbf=SVC()
svc_rbf.fit(x_train,y_train)
p=svc_rbf.predict(x_test)
svc_rbf.score(x_train,y_train)
print(f"accuracy score is {accuracy_score(y_test,p)*100}") 
print("cvscore",cross_val_score(svc_rbf,x,y,cv=5).mean())
print('\n Classification Report:\n', classification_report(y_test,p))
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, recall_score, classification_report
Gb=GradientBoostingClassifier()
Gb.fit(x_train,y_train)
p=Gb.predict(x_test)
print(f"accuracy score is {accuracy_score(y_test,p)*100}")
print("cvscore",cross_val_score(Gb,x,y,cv=5).mean())
print('\n Classification Report:\n', classification_report(y_test,p))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score, classification_report
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
p=dt.predict(x_test)
print(f"accuracy score is {accuracy_score(y_test,p)*100}")
print("cvscore",cross_val_score(dt,x,y,cv=5).mean())
print('\n Classification Report:\n', classification_report(y_test,p))
```
